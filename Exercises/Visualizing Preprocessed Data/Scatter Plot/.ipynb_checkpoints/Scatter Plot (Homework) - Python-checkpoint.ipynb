```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_student_scores = pd.read_csv("student_scores_data.csv")
```

```python
df_student_scores
```

```python
plt.figure(figsize = (12,8))
scatter = plt.scatter(df_student_scores['SAT'],
            df_student_scores['GPA'],
            alpha = 0.6,
            c = df_student_scores['SAT'],
            cmap = 'mako')

plt.legend(*scatter.legend_elements(),
          loc = "upper left",
          title = "Student's SAT"
          )

plt.title("Relationship between SAT and GPA of Students",
          fontsize = 14, fontweight = 'bold')

plt.xlabel("SAT of students", weight = 'bold')
plt.ylabel("GPA of students" , weight = 'bold')
plt.show()
```

```python
plt.savefig("Relationship between SAT and GPA of Students")
```
