```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize = (9,6))
#Change the size of the bar chart

plt.bar(x = df_used_cars['Brand'], height = df_used_cars['Cars Listings'], color = "midnightblue")
# x used for the brand names / horizontal axis and height for y axis / car listings. 
# Separate each color string as the conventional 'rgbwymc' is not accepted by python.

plt.xticks(rotation = 45, fontsize = 13)
#Method used to rotate the labels on the x axis and avoid overlapping.

plt.yticks(fontsize = 13)
#Change the font size of y-axis labels.

plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold")
#Specify the title for the bar chart.

plt.ylabel("Number of Listings", fontsize = 13)
#Specify the y axis label for the bar chart.

plt.show()
#used to show the plot

plt.savefig("Used Cars Bar.png")
```
