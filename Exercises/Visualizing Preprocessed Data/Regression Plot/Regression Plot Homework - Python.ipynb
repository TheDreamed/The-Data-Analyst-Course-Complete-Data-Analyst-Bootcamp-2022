```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_insurance_sweden = pd.read_csv("regression_auto_insurance_sweden.csv")
df_insurance_sweden
```

```python
#plt.figure(figsize = (10,8))
sns.set(rc = {'figure.figsize': (9,6)})
sns.regplot(x = 'Claims',
            y = 'Payment',
            data = df_insurance_sweden,
            scatter_kws = {'color': 'k'},
            line_kws = {'color': 'red'})

plt.xlabel("Number of Claims")
plt.ylabel("Number of Payments")
plt.title("Effects of Claims to the Number of Payments on Auto Insurance in Sweden", fontsize = 14, weight = 'bold')
plt.show()
```
