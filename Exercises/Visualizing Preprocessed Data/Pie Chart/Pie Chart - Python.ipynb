```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_type = pd.read_csv("pie_chart_data.csv")
```

```python
df_fuel_engine_type
sns.set_palette('colorblind')
```

```python
plt.figure(figsize = (10,8))
#Change the size of the bar chart

plt.pie(df_fuel_engine_type['Number of Cars'],
        labels = df_fuel_engine_type['Engine Fuel Type'].values,
        autopct = '%.2f%%',
        textprops = {'size' : 'x-large', 
                     'fontweight' : 'bold', 
                     'rotation' : '30',
                     'color': 'w'})
plt.legend()
plt.title('Cars by engine Fuel Type', fontsize = 18, fontweight = 'bold')
#Method used to use a pie chart.
plt.show()
```

```python
plt.savefig("Engine Fuel Type Pie Chart.png")
```
