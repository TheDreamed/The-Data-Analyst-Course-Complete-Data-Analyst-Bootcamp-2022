```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

#This chart is used when we have time data.
#Comparing volume among categories
#Atleast three categories.
#Ordered and numerical.
#No more than 6-7 categories
#Y axis needs to start @ 0
```

```python
df_fuel_engine_types = pd.read_csv("stacked_area_chart_data.csv")
```

```python
df_fuel_engine_types
```

```python
colors = ['#011638', '#7e2987', 'ef2026']
#Create a list of colors for our area chart.

labels = ['Diesel', 'Petrol', 'Gas']
#Create a list of labels for our area chart.

sns.set_style('white')
#Used to change the background color of the chart.

plt.figure(figsize = (12,6))
#Change the size of our area chart

plt.stackplot(df_fuel_engine_types["Year"],
             df_fuel_engine_types["Diesel"],
             df_fuel_engine_types["Petrol"],
             df_fuel_engine_types["Gas"],
             colors = colors,
             edgecolor = 'none')

plt.legend(labels, loc = "upper left")
#Assign colors list to the colors argument of the stackplot method.

plt.xticks(df_fuel_engine_types["Year"], rotation = 45)
#Rotate the Year labels to avoid overlapping.

plt.ylabel("Number of Cars", fontsize = 13, fontweight = 'bold')
plt.title("Popularity of Engine Fuel Types (1982 -2016)", fontsize = 14, weight = "bold")

sns.despine()
#Used to remove the top and right border of the chart.
plt.show()
```
