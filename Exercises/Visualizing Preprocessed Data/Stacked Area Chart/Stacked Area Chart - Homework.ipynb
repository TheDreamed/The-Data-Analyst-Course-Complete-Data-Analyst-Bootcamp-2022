```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
chart = pd.read_csv("area_chart_homework_data.csv")
chart
```

```python
chart.drop(columns=['Unnamed: 4', 'Unnamed: 5'])
```

```python
chart = chart.drop(columns=['Services'])
```

```python
chart
```

```python
colors = ['#011638', '#7e2987']
#Create a list of colors for our area chart.

labels = ['Domestic', 'International']
#Create a list of labels for our area chart.

sns.set_style('white')
#Used to change the background color of the chart.

plt.figure(figsize = (10,5))
#Change the size of our area chart

plt.stackplot(chart["Date"],
             chart["Domestic"],
             chart["International"],
             colors = colors,
             edgecolor = 'none')

plt.legend(labels, loc = "upper left")
#Assign colors list to the colors argument of the stackplot method.

plt.xticks(chart["Date"], rotation = 45)
#Rotate the Year labels to avoid overlapping.

plt.ylabel("International and Domestic", fontsize = 13, fontweight = 'bold')
plt.title("2000 Services Comparison Domestic vs International 2020", fontsize = 14, weight = "bold")

sns.despine()
#Used to remove the top and right border of the chart.
plt.show()
```
